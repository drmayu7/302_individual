```python
#Demonstrate conversion of the horse-colic.txt file to horse-colic.csv - pandas -- second method
import csv

with open('dataset/horse-colic.txt','r')as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(",") for line in stripped if line)
    with open('dataset/horse-colic2.csv','w')as out_file:
        writer = csv.writer(out_file)
        writer.writerows(lines)
```

```python
#Read and evaluate the following code in Figure 1. What method is the code using to reduce the missing value from 1605 to 0? Justify your answer.
print ('Method: Mean Imputation of Missing Values (Univariate)')
print ('Justification : The Scikit Learn - Imputer Module is able to reduce null values to 0 by replacing the null values with a predicted values based on columns and rows pattern of the dataset')

from numpy import isnan
from pandas import read_csv
from sklearn.impute import SimpleImputer

# load dataset
# replace the '?' with NaN

#convert dataframe to list-array
data = horse_colic_final.values

#remove outcome column from dataset
totalCol = data.shape[1]
ix = [i for i in range(totalCol) if i !=23]
X = data[:,ix]
y = data[:,23]

#count and display number of missing values in pre-imputated dataset
print('Missing: %d' % sum(isnan(X).flatten()))

#using mean imputation from scikit-lean SimpleImputer - calculate mean within columns and rows of missing values and point the intersection of those values
imputer = SimpleImputer(strategy='mean')
#predict missing values based on mean of its column - imputer
imputer.fit(X)
Xtrans = imputer.transform(X)

#count and display number of missing values in imputated dataset
print('Missing: %d' % sum(isnan(Xtrans).flatten()))
```
